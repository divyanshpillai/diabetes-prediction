```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**Data Collection and Analysis**

```python
# Loading the dataset to a pandas dataframe
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
# Group the data seperately of diabetic and non-diabetic people
diabetes_dataset.groupby('Outcome').mean()
```

**Seperating the data and Outcome column**

```python
X = diabetes_dataset.drop(columns='Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
X
```

```python
Y
```

**Data Standardization**

```python
# Standardized the data using StandardScaler function that we import from sklearn.preprocessing
scaler = StandardScaler()
```

**One way to fit and transform the data**

```python
# Fit and transform the data
standardized_scaler = scaler.fit_transform(X)
```

```python
X = standardized_scaler
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

**Train Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Train the Model**

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

**Model Evaluation**

Accuracy Score

```python
# Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on training data', training_data_accuracy)
```

```python
# Accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test data', test_data_accuracy)
```

Making a Predictive System

```python
input_data = (11,143,94,33,146,36.6,0.254,51)

# Changing the input data into numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array as we are predicting for one example
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0] == 0):
  print('The person is non-diabetic')
else:
  print('The person is diabetic')
```
